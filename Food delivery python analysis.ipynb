```python
food.drop(columns=['Unnamed: 12'], inplace=True)

```

```python
food.info()
```

```python
# Map the 'Monthly Income' to categorical variables
income_categories = {
    'No Income': 'No Income',
    'Below Rs.10000': 'Low Income',
    '10001 to 25000': 'Medium Income',
    '25001 to 50000': 'Upper-Medium Income',
    'More than 50000': 'High Income'
}
food['Income Category'] = food['Monthly Income'].map(income_categories)
```

```python
food.head()
```

```python
food['Output'].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization 1: Bar Plot for Monthly Income Categories
plt.figure(figsize=(10, 6))
income_order = ['No Income', 'Low Income', 'Medium Income', 'Upper-Medium Income', 'High Income']
sns.countplot(x='Income Category', data=food, order=income_order, palette='Set2')
plt.title('Distribution of Monthly Income Categories')
plt.xlabel('Income Category')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 2: Histogram for Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(food['Age'], bins=30, kde=False, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 3: Count Plot for Marital Status
plt.figure(figsize=(10, 6))
sns.countplot(x='Marital Status', data=food, palette='Set1')
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 4: Count Plot for Gender
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=food, palette='Set3')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 5: Pie Chart for Educational Qualifications
plt.figure(figsize=(10, 6))
education_counts = food['Educational Qualifications'].value_counts()
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Educational Qualifications')
plt.show()
```

```python
# Visualization 6: Bar Plot for Feedback
plt.figure(figsize=(10, 6))
sns.countplot(x='Feedback', data=food, palette='coolwarm')
plt.title('Feedback Distribution')
plt.xlabel('Feedback')
plt.ylabel('Count')
plt.show()
```
